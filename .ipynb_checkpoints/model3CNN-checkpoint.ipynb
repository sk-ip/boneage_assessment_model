```python
def DataPreprocessor(folder):
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename),0)
        resize = cv2.resize(img, (width, height))
        norm = cv2.normalize(resize, normalized, 0, 128, cv2.NORM_MINMAX)
        img = cv2.resize(norm, dim, interpolation = cv2.INTER_AREA)
        cv2.imwrite(os.path.join(save_dir, filename), img)
        print('Image:',filename,'processed and saved.')

DataPreprocessor(train_dir)
```

```python
# reading the data from csv file
ids = []
boneage = []
male = [] #1 for yes, 0 for false
with open('boneage-training-dataset.csv') as csvfile:
    content = csv.reader(csvfile, delimiter=',')
    for row in content:
        i = row[0]
        age = int(row[1])
        if row[2]=='True':
            m=(1)
        else:m=(0)
        
        ids.append(i)
        boneage.append(age)
        male.append(m)
        
```

```python
# creating a dictionary id:age,male(o or 1)
data={}
for i in range(len(ids)):
    data[ids[i]] = [boneage[i], male[i]]
    
print(data)
```

```python
#when storing values of a dixtionary use pickling
#Pickling is used to store python objects. This means things like lists, dictionaries, class objects, and more

# saving data using pickle
pickle_out=open("dict.pickle","wb")
pickle.dump(data, pickle_out)
pickle_out.close()

# reading the data using pickle
#pickle_in = open("dict.pickle","rb")
#x = pickle.load(pickle_in)
#pickle_in.close()

#np.save('data.npy', data)
#data = np.load('data.npy')
```

```python
def SplitData(folder, file_to_save_to):
    pickle_in=open("dict.pickle","rb")
    data=pickle.load(pickle_in)
    pickle_in.close()
    data_dict={}
    for filename in os.listdir(folder):
        data_dict[filename.replace('.png','')]=data[filename.replace('.png','')]
    pickle_out=open("{0}.pickle".format(file_to_save_to),"wb")
    pickle.dump(data_dict, pickle_out)
    pickle_out.close()
    print('data dictionary saved as:{0}.pickle'.format(file_to_save_to))
```

```python
SplitData(r'C:\Users\Sudhanshu\bone_age_model\test',"test")
SplitData(r'C:\Users\Sudhanshu\bone_age_model\train',"train")
SplitData(r'C:\Users\Sudhanshu\bone_age_model\valid',"valid")
```
